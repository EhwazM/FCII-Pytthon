```python
import numpy as np
from numpy.random import uniform as uni
import matplotlib.pyplot as plt

f = lambda x: 1/(np.exp(x) + 1)

def Transform(a, b, u, f):
    x = (1/u) - 1
    return ((1/u**2) * f(x))

a, b = (0,1e2028)
n = 1000

yRandom = uni (0, 1, n)
functionValue = Transform(a,b,yRandom, f)
integral = np.mean(functionValue)

print(integral)
```
