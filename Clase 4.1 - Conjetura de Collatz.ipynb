```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Digite un número entero cualqueira:"))
listCollatz = [n]

while not(n == 1):
    
    if n%2 == 0:

        n = int(n / 2)

    else:

        n = int(3*n + 1)

    listCollatz.append(n)

print(listCollatz)

x = np.arange(len(listCollatz))

plt.plot(x, listCollatz)
plt.xlabel("N° iteraciones")
plt.ylabel("Serie de Collatz")
plt.title("Serie de Collatz vs N° iteraciones")
plt.show()
        
```
