```python
import numpy as np
import matplotlib.pyplot as plt

# y' + 2y = 0
# exp(-2t)

def EulerExplicit(y0, rng, h, number, f):

    y_v = np.zeros(number + 1)
    y_v[0] = y0
    
    for i in range(number):
        y_v[i+1] = y_v[i] + (h * f(rng[i], y_v[i]))
    
    return y_v
```

```python
y = lambda t: np.exp(-2 * t)
f = lambda t, y: -2*y

t0 = 0
tf = 5
h = 1e-1
n = int((tf-t0)/h)

test = lambda t: np.sin(t)
t = np.linspace(t0, tf, n+1)

y_values = EulerExplicit(1, t, h, n, f)

fig, ax = plt.subplots(1, 2, figsize = (10, 3.5), layout = "constrained")

ax[0].plot(t, y_values)
ax[0].plot(t, y(t), ls = "--")
ax[0].set_xlabel("t")
ax[0].set_ylabel("y")

error = np.abs(y(t) - y_values)

ax[1].plot(t, error)
ax[0].grid()
ax[0].legend(["hola", "hola"])
ax[1].grid()
plt.show()
```

```python
h_medio = h/2
n_m = int((tf-t0)/h_medio)
t_m = np.linspace(t0, tf, n_m + 1)
y_values_m = EulerExplicit(1, t_m, h_medio, n_m, f)

fig, ax2 = plt.subplots(1, 2, figsize = (10, 3.5), layout = "constrained")

ax2[0].plot(t, y_values_m)
ax2[0].plot(t, y(t), ls = "--")
ax2[0].set_xlabel("t")
ax2[0].set_ylabel("y")

error = np.abs(y(t) - y_values_m)

ax2[1].plot(t, error)
ax2[0].grid()
ax2[0].legend(["hola", "hola"])
ax2[1].grid()
plt.show()
```
