```python
import numpy as np
import matplotlib.pyplot as plt

# A = np.array([[3, -0.1, -0.2],
#              [0.1, 7, -0.3],
#              [0.3, -0.2, 10]])

# B = np.array([[7.85],
#              [-19.3],
#              [71.4]])

A = np.array([[4, 2, 5],
             [2, 5, 8],
             [5, 4, 3]])

B = np.array([[60.7],
             [92.9],
             [56.3]])
tol = 1e-4
max_iter = 1e2
n,m = A.shape
X = np.zeros(n)
dif = np.ones(n)
error = 2 * tol

iter = 0

while not(error < tol or iter > max_iter):
    for i in range(n):
        sum = 0
        for j in range(m):
            if j != i:
                sum += A[i,j]*X[j]
        new = (B[i] - sum) / A[i,i]
        dif[i] = np.abs(new - X[i])
        X[i] = new
    error = np.max(dif)
    iter += 1

X = np.transpose([X])

print(X)
```
