```python
import numpy as np 
from numpy.random import uniform as uni
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)

a, b = 0, np.pi
n = 10000
x = uni(a, b, n)
f_max = np.max(f(x))
y = uni(0, f_max, n)

suma = 0

for i in range(n):
    if y[i] <= f(x[i]):
        suma += 1
        
integral = (b-a) * f_max * (suma/n)
# integral = 0

# m = 1000

# for i in range(m):
#     integral += (b-a) * f_max * (suma/n)

# integral = integral / m

print(integral)
```

```python
num_muestras = 1000
l_sup = np.pi
l_inf = 0
sum = 0

x = uni(l_inf, l_sup, num_muestras)

for i in range(num_muestras):
    sum += f(x[i])

integral = ((l_sup - l_inf) * sum) / (num_muestras)

print(integral)
```
