```python
import numpy as np 
from numpy.random import uniform as uni
import matplotlib.pyplot as plt
from sympy import *

def f(x, y):
    return np.sin(x) * np.sin(y)

a, b = 0, np.pi
n = 10000
x = uni(a, b, n)
y = uni(a, b, n)

f_max = np.max(f(x,y))

z = uni(0, f_max, n)

suma = 0

for i in range(n):
    if z[i] <= f(x[i], y[i]):
        suma += 1
        
integral = (b-a)**2 * f_max * (suma/n)

print(integral)
```
