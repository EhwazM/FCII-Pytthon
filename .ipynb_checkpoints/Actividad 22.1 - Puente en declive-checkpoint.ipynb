```python
import numpy as np
import matplotlib.pyplot as plt

def FFT(signal):
    N = len(signal)
    if N <= 1:
        return signal
    even = FFT(signal[0::2])
    odd = FFT(signal[1::2])
    spectra = np.zeros(N, dtype = "complex")

    for k in range(N//2):
        angle = -2*np.pi*k/N
        factor = odd[k]*complex(np.cos(angle), np.sin(angle))
        spectra[k] = even[k] + factor
        spectra[k+(N//2)] = even[k] - factor

    return spectra

def FFTFrec(N, d):
    frecuencies = []

    for i in range(N):
        if i <= N//2:
            freq = i/(N*d)
            
        else:
            freq = (i - N)/(N*d)

        frecuencies.append(freq)

    return frecuencies
```
