```python
import numpy as np
import matplotlib.pyplot as plt

A1 = np.array([[4, 2, 5],
             [2, 5, 8],
             [5, 4, 3]])

A2 = np.array([[3, -0.1, -0.2],
             [0.1, 7, -0.3],
             [0.3, -0.2, 10]])

A3 = np.array([[2, 3, 1],
             [4, 7, 2],
             [6, 18, 5]])

B = np.array([[60.7],
             [92.9],
             [56.3]])
```

```python

def Decomposer(matrix, vector):
    n,m = matrix.shape
    matrix0 = np.copy(matrix)
    
    L = np.zeros((n,m), dtype = 'float')
    U = np.copy(matrix0)
    Y = np.zeros(n, dtype='float')
    X = np.zeros(n, dtype='float')
    
    for i in range(n):
        for j in range(m): 
            if i == j:  
                L[i,j] = 1
                
            if i < j:
                factor = matrix0[i,j] / matrix0[i,i]
                L[i,j] = factor
                
                for c in range(n):
                    matrix0[j,c] = matrix0[j,c] - factor * matrix0[i,c]
                    U[j,c] = matrix0[j,c]
                    
    for i in range(n):
        sum = 0

        for j in range(n):
            sum += L[i,j]*Y[j]

        Y[i] = (vector[i] - sum) / L[i,i]

    for i in range(n-1, -1, -1):
        sum = 0

        for j in range(n-1, -1, -1):
            sum += U[i,j]*X[j]

        X[i] = (Y[i] - sum) / U[i,i]

    det = 1

    for i in range(n):    
        det *= U[i,i]
        
    return L, U, X, Y, det
    
results = Decomposer(A3, B)

print(results[0])
print(results[1])
print('\n')
print(np.dot(results[0],results[1]))
print(A3)
print('\n')
print(results[2])
print(results[3])
print('\n')
print(B)
print(np.transpose([np.dot(A1, results[2])]))
print('\n')
print(results[4])
```
