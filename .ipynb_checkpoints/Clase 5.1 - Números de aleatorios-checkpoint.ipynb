```python
import numpy as np
import matplotlib.pyplot as plt

def LinealGenerator(mod, mult, incr, seed):
    while True:
        yield seed
        seed = (((mult * seed) + incr)  % mod)

m = 48
a = 25
c = 19
x = 47

random = []
number = LinealGenerator(m, a, c, x)

for i in range(50):

    random = np.append(random, next(number))

print(random)

    
```
