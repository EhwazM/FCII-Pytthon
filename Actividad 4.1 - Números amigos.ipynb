```python
import numpy as np
import matplotlib.pyplot as plt

def friendsNumbers(p,q, divp, divq):

    i = p
    j = q

    sump = 0
    sumq = 0

    while i>1:

        i -= 1

        if p%i == 0:
            
            divp.append(i)
            sump += i 

    while j>1:

        j -= 1

        if q%j == 0:
            
            divq.append(j)
            sumq += j

    if sumq == p and sump == q:
        print("\n Los números",p,"y",q,"son amigos.\n")

    else:
        print("\n No se conocen ni en pintura.\n")

    return divp, divq, sump, sumq
    
p = int(input(""))
q = int(input(""))

divp = []
divq = []

results = friendsNumbers(p, q, divp, divq)

print(" La lista de divisores de p es:", results[0])
print("\n La lista de divisores de q es:", results[1])

print("\n La suma de los divisores de p es:", results[2]," y la de los divisores de q es:", results[3])

#La cantidad de \n es para que se vea lindo el output.

```
