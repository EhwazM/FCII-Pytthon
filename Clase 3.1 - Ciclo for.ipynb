```python
import numpy as np
import matplotlib.pyplot as plt

n = 10 # rows
m = 10 # colums

casillero = np.zeros(shape = (n,m), dtype = int) # or just (n,m) without shape parameter. dtype is just for simplicity and problem conditions

option = 0

while not(option == 5):
    
    print("\n1. Asiganar casillero.")
    print("2. Devolver casillero.")
    print("3. Estadísticas.")
    print("4. Mostrar casilleros.")
    print("5. Salir del Menú.\n")
    
    option = int(input("\nDigite el número que correpende a su petición: \n"))
    
    if option == 1:
        
        print("Escriba la fila y la columna del casillero deseado (si no hay disponibilidad se le avisará)")
        row = int(input("Escoja la fila:"))
        column = int(input("Escoja la columna:"))
        id = int(input("Digite su número de identificación estudiantil:"))
    
        if casillero[row - 1, column - 1] == 0 and id > 0:
            
            casillero[row - 1, column - 1] = id
            print("Su casillero fue asignado.")
            
        elif casillero[row - 1, column - 1] != 0 and id > 0:
    
            print("Su casillero está ocupado, elija otro por favor.")
    
        else:
            print("Digite un número de identificación válido.")
            
    elif option == 2:
    
        print("Escriba la fila y la columna del casillero que se desea devolver\n")
        row = int(input("Escoja la fila:\n"))
        column = int(input("Escoja la columna:\n"))
        id = int(input("Digite su número de identificación estudiantil:\n"))
    
        if casillero[row - 1, column - 1] == id and id > 0:
            
            casillero[row - 1, column - 1] = 0
            print("Casillero devuelto")
            
        elif casillero[row - 1, column - 1] != id and id > 0:
            
            print("La identificación ingresada no coincide con la del casillero, vuelva a intentar.")
        
        else:
            print("Digite un número de identificación válido.")
    
    elif option == 3:

        passwd = int(input("Introduzca su contraseña de administrador:\n"))
        
        available = 0
        occupied = 0

        if passwd == 202428:
        
            for i in range(n):
                for j in range(m):
                    
                    if casillero[i,j] == 0:
                        
                        available += 1
                        
                    else:
        
                        occupied += 1
            print("El número de casilleros disponibles es de:", available)
            print("El número de casilleros ocupados es de:", occupied)
        
        else:

            print("Contraseña incorrecta.")
    
    elif option == 4:
    
        print(casillero)
```
